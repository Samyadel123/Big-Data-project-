```python
from minio import Minio
clinet = Minio(
    endpoint="localhost:9000",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)
```

```python
buckets_list = clinet.list_buckets()
for bucket in buckets_list:
    print(bucket.name, bucket.creation_date)
```

```python
# function to check if bucket exists
def bucket_exists(bucket_name):
    try:
        clinet.stat_bucket(bucket_name)
        return True
    except Exception as e:
        return False
# write content to bucket 
def write_to_bucket(bucket_name:str,
                    file_path:str):
    if not bucket_exists(bucket_name):
        clinet.make_bucket(bucket_name)
    with open(file_path, 'rb') as file_data:
        clinet.put_object(bucket_name, file_path, file_data, -1)
```
